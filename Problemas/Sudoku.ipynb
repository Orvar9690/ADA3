```python
import time

# Resolver un Sudoku con backtracking
def resolver_sudoku(tablero):
    """
    Resuelve un tablero de Sudoku utilizando backtracking.
    Args:
        tablero (list[list[int]]): Matriz de 9x9 con el Sudoku a resolver.
    Returns:
        bool: True si se encuentra solución, False si no.
    """
    vacio = encontrar_vacio(tablero)
    if not vacio:
        return True  # Si no hay celdas vacías, el tablero está resuelto
    fila, col = vacio

    for num in range(1, 10):  # Intentar números del 1 al 9
        if es_valido(tablero, num, (fila, col)):
            tablero[fila][col] = num  # Asignar el número

            if resolver_sudoku(tablero):
                return True  # Si se encuentra solución, retornar

            tablero[fila][col] = 0  # Deshacer si no es solución

    return False  # No se encontró solución


def encontrar_vacio(tablero):
    """
    Encuentra la primera celda vacía en el tablero.
    Args:
        tablero (list[list[int]]): Matriz de Sudoku.
    Returns:
        tuple: (fila, col) de la celda vacía, o None si no hay.
    """
    for i in range(len(tablero)):
        for j in range(len(tablero[0])):
            if tablero[i][j] == 0:
                return (i, j)  # Retorna la posición de la celda vacía
    return None


def es_valido(tablero, num, posicion):
    """
    Verifica si un número puede colocarse en una posición específica.
    Args:
        tablero (list[list[int]]): Matriz de Sudoku.
        num (int): Número a verificar.
        posicion (tuple): (fila, col) de la celda.
    Returns:
        bool: True si es válido, False si no.
    """
    fila, col = posicion

    # Verificar fila
    if num in tablero[fila]:
        return False

    # Verificar columna
    if num in [tablero[i][col] for i in range(len(tablero))]:
        return False

    # Verificar subcuadrícula
    sub_fila = (fila // 3) * 3
    sub_col = (col // 3) * 3
    for i in range(sub_fila, sub_fila + 3):
        for j in range(sub_col, sub_col + 3):
            if tablero[i][j] == num:
                return False

    return True


def imprimir_tablero(tablero):
    """
    Imprime el tablero de Sudoku en formato legible.
    Args:
        tablero (list[list[int]]): Matriz de Sudoku.
    """
    for i in range(len(tablero)):
        if i % 3 == 0 and i != 0:
            print("-" * 21)  # Separador de bloques

        for j in range(len(tablero[0])):
            if j % 3 == 0 and j != 0:
                print("| ", end="")

            if j == 8:
                print(tablero[i][j])
            else:
                print(str(tablero[i][j]) + " ", end="")


# Ejemplo de uso
tablero = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9],
]

print("Sudoku inicial:")
imprimir_tablero(tablero)

start_time = time.time()
if resolver_sudoku(tablero):
    print("\n¡Sudoku resuelto!")
    print("\nSe resolvió en: ",time.time() - start_time)
    imprimir_tablero(tablero)
else:
    print("\nNo se encontró solución.")
```
